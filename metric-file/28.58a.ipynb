```python
reset()
from SageNP import NewmanPenrose
MyManifold = Manifold(4 , 'MyManifold', r'\mathcal{Man}')
MyCoordinates.<u,r,x,y> = MyManifold.chart(r' u r x y ')
var('m')
var('k')

Q=function('Q',imag_part_func=0)(x,y,u)
A=function('A',imag_part_func=0)(x,y,u)
B=(2*m+2*k*A)/r-k*Q*Q/(2*r^2)

lvec=[1, 0, 0, 0]
nvec=[-B/2, 1, 0, 0]
mvec=[0, 0, r/sqrt(2),i*r/sqrt(2)]
mbarvec=[0,0,r/sqrt(2),-i*r/sqrt(2)]



tetrad_type = 'covariant'  # 'covariant'4ya da 'contravariant' olabilir.
########################################/2###############################################################
f=MyManifold.scalar_field(function('f')(*MyCoordinates))
# Define the object "metric" of the class "SageNP":
met = NewmanPenrose(MyManifold, MyCoordinates, lvec, nvec, mvec, mbarvec, tetrad_type)


#########################################################################################################


#NP elemanlarını hesaplatıyoruz :
met.calculate_spincoefficients()

met.calculate_Weyl()

met.calculate_Ricci()

met.calculate_NPeq()

met.calculate_Bianchi()

met.Petrov_fromWeyl()


#Sonuçları görmek isterseniz aşağıdaki kodu kullanabiliriz :
#met.show_allNP()

#Bu kod latex dosyasını düzenlemek için eklenmiştir. (herhangi bir değişiklik yapmayın!)
def fitEq(expr):
    expanded_expr = expr.expand()

    latex_expr = latex(expanded_expr)

    return latex_expr

#Dönüşümleri Hesaplatıyoruz :
var('z')
met.type_A_transformation(z)
met.type_B_transformation(z)
met.type_C_transformation(z)


#Dönüşümleri Görmek İçin Kullanabiliriz :
#met.show_type_A_transformation()
#show('****************************************************')
#met.show_type_B_transformation()
#show('****************************************************')
#met.show_type_C_transformation()

# Yardımcı fonksiyonlar
def compare_expressions(orig, trans):
    """Dönüşümleri karşılaştır"""
    if trans == 0:
        return "0"
    try:
        if bool(trans == orig):
            return None
    except:
        if str(trans) == str(orig):
            return None
    return str(latex(fitEq(trans)))

def write_transformation_section(fout, coeffs, title):
    """Dönüşüm bölümünü yaz"""
    fout.write(f"\n\\textit{{{title}:}}\n\n")

    for symbol, orig, trans in coeffs:
        result = compare_expressions(orig, trans)
        if result is None:
            fout.write(f"\\begin{{dmath}}\n{symbol}' = {symbol}\n\\end{{dmath}}\n\n")
        else:
            fout.write(f"\\begin{{dmath}}\n{symbol}' = {result}\n\\end{{dmath}}\n\n")

def write_transformations(fout, met, transform_type):
    """Tüm dönüşümleri yaz"""
    tr_suffix = f"_tr{transform_type}"

    fout.write(f"\n\\newpage\n\\underline{{{transform_type} Type transformations:}}\n")

    # Spin coefficients
    spin_coeffs = [
        ('\\rho', met.rhoNP.expr(), getattr(met, f'rhoNP{tr_suffix}')),
        ('\\sigma', met.sigmaNP.expr(), getattr(met, f'sigmaNP{tr_suffix}')),
        ('\\kappa', met.kappaNP.expr(), getattr(met, f'kappaNP{tr_suffix}')),
        ('\\mu', met.muNP.expr(), getattr(met, f'muNP{tr_suffix}')),
        ('\\lambda', met.lambdaNP.expr(), getattr(met, f'lambdaNP{tr_suffix}')),
        ('\\nu', met.nuNP.expr(), getattr(met, f'nuNP{tr_suffix}')),
        ('\\alpha', met.alphaNP.expr(), getattr(met, f'alphaNP{tr_suffix}')),
        ('\\beta', met.betaNP.expr(), getattr(met, f'betaNP{tr_suffix}')),
        ('\\tau', met.tauNP.expr(), getattr(met, f'tauNP{tr_suffix}')),
        ('\\pi', met.piNP.expr(), getattr(met, f'piNP{tr_suffix}')),
        ('\\epsilon', met.epsilonNP.expr(), getattr(met, f'epsilonNP{tr_suffix}')),
        ('\\gamma', met.gammaNP.expr(), getattr(met, f'gammaNP{tr_suffix}'))
    ]
    write_transformation_section(fout, spin_coeffs, "Spin coefficients")

    # Weyl components
    weyl_comps = [
        ('\\Psi_0', met.Psi0NP.expr(), getattr(met, f'Psi0NP{tr_suffix}')),
        ('\\Psi_1', met.Psi1NP.expr(), getattr(met, f'Psi1NP{tr_suffix}')),
        ('\\Psi_2', met.Psi2NP.expr(), getattr(met, f'Psi2NP{tr_suffix}')),
        ('\\Psi_3', met.Psi3NP.expr(), getattr(met, f'Psi3NP{tr_suffix}')),
        ('\\Psi_4', met.Psi4NP.expr(), getattr(met, f'Psi4NP{tr_suffix}'))
    ]
    write_transformation_section(fout, weyl_comps, "Weyl components")

    # Ricci components
    ricci_comps = [
        ('\\Phi_{00}', met.Phi00NP.expr(), getattr(met, f'Phi00NP{tr_suffix}')),
        ('\\Phi_{01}', met.Phi01NP.expr(), getattr(met, f'Phi01NP{tr_suffix}')),
        ('\\Phi_{02}', met.Phi02NP.expr(), getattr(met, f'Phi02NP{tr_suffix}')),
        ('\\Phi_{10}', met.Phi10NP.expr(), getattr(met, f'Phi10NP{tr_suffix}')),
        ('\\Phi_{11}', met.Phi11NP.expr(), getattr(met, f'Phi11NP{tr_suffix}')),
        ('\\Phi_{12}', met.Phi12NP.expr(), getattr(met, f'Phi12NP{tr_suffix}')),
        ('\\Phi_{20}', met.Phi20NP.expr(), getattr(met, f'Phi20NP{tr_suffix}')),
        ('\\Phi_{21}', met.Phi21NP.expr(), getattr(met, f'Phi21NP{tr_suffix}')),
        ('\\Phi_{22}', met.Phi22NP.expr(), getattr(met, f'Phi22NP{tr_suffix}'))
    ]
    write_transformation_section(fout, ricci_comps, "Ricci components")

# Ana latex dosyası için dönüşüm bölümlerini yazan fonksiyon
def write_all_transformations(fout, met):
    for transform_type in ['A', 'B', 'C']:
        write_transformations(fout, met, transform_type)




# Latex formatında rapor dosyası oluşturmak için bu kodu kullanabilirsiniz.
# (Dosyanin adini uygun sekilde degistirmeyi unutmayin.)
fout = open('dosya4.tex',"w")
tmpstr=r'''\documentclass[landscape]{article}
%\documentclass[11pt,a4paper,reqno]{amsart}
%%%%%%%%%%%%%%%%%%%%%%
\usepackage[fleqn]{amsmath} %Denklemleri sola daya
\usepackage{amssymb}
\usepackage{authblk}
\usepackage{caption}
\usepackage{cite}
\usepackage{xcolor}
\usepackage{color}
\usepackage{float}
\usepackage{geometry}
\usepackage{graphicx}
\usepackage{mdframed}
\usepackage{hyperref}
\usepackage{textcomp}
\usepackage{epstopdf}
\usepackage{subcaption}
\usepackage{lscape}
\usepackage{breqn}
%\usepackage{soul}
%%%%%%%%%%%%%%%%%%%%%%
\setlength\parindent{0pt}
\setlength{\parskip}{5pt}
\geometry{left=20mm,right=20mm,top=30mm,bottom=20mm}
\hyphenation{Schwarz-schild}
%%%%%%%%%%%%%%%%%%%%%%
\newcommand{\etalchar}[1]{$^{#1}$}
\newcommand\hl[1]{%
        \bgroup
        \hskip0pt\color{red!80!black}%
        #1%
        \egroup
}
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
\begin{document}
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
\textbf{Newman-Penrose Formalism Calculations for}
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\vspace{10pt}
\textbf{Metric:}

\textit{Name:García D. and Salazar I.}



\textit{Reference(s):Exact Solutions to Einstein's Field Equations Second Edition Equation: 24.22}

\textit{Line element:}

\begin{dmath}'''
fout.write(tmpstr)


fout.write("ds^2 " + str(latex(met.g.display()))[2:])
fout.write(r"\end{dmath}")


fout.write(r"\textit{Coordinates:} "+str(latex(MyCoordinates[:])))
tmpstr=r'''
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\vspace{10pt}
\textbf{Null tetrad components: }

\textit{Covariant components:}
\begin{dmath}
'''
fout.write(tmpstr)

fout.write(r"l&= " +r"\left("+str(latex(met.lNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.lNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.lNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.lNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"n&= " +r"\left("+str(latex(met.nNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.nNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.nNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.nNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)

fout.write(r"m&= " +r"\left("+str(latex(met.mNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.mNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.mNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.mNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)

fout.write(r"\overline{m}&= " +r"\left("+str(latex(met.mbarNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.mbarNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.mbarNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.mbarNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")

tmpstr=r'''
\end{dmath}

\textit{Contravariant components:}
\begin{dmath}
'''
fout.write(tmpstr)

fout.write(r"l&= " +r"\left("+str(latex(met.lupNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.lupNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.lupNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.lupNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"n&= " +r"\left("+str(latex(met.nupNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.nupNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.nupNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.nupNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)

fout.write(r"m&= " +r"\left("+str(latex(met.mupNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.mupNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.mupNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.mupNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)

fout.write(r"\overline{m}&= " +r"\left("+str(latex(met.mbarupNP[0])) +r"\right) d"+str(latex(MyCoordinates[0]))  +r"+\left( " +str(latex(met.mbarupNP[1])) +r"\right) d"+str(latex(MyCoordinates[1])) + r"+\left( " + str(latex(met.mbarupNP[2])) +r"\right) d"+str(latex(MyCoordinates[2]))
           +r"+\left( " +str(latex(met.mbarupNP[3])) +r"\right) d"+str(latex(MyCoordinates[3]))+r" \\ ")

tmpstr=r'''
\end{dmath}


\newpage
\textbf{Directional derivatives: }

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"D&=" + str(latex(fitEq(met.DlNP(f).expr()))) +r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"\Delta&=" + str(latex(fitEq( met.DeltanNP(f).expr()))) +r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"\delta&=" + str(latex(fitEq( met.deltamNP(f).expr()))) +r" \\ ")
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"\overline{\delta}&=" + str(latex(fitEq( met.deltambarNP(f).expr()))))
tmpstr=r'''
\end{dmath}

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\newpage
\textbf{Spin coefficients: }

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\rho&="+str(latex(fitEq(met.rhoNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\sigma&="+str(latex(fitEq(met.sigmaNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\kappa&="+str(latex(fitEq(met.kappaNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\mu&="+str(latex(fitEq(met.muNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\lambda&="+str(latex(met.lambdaNP.expr()))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\nu&="+str(latex(fitEq(met.nuNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\alpha&="+str(latex(fitEq(met.alphaNP.expr()))))

tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\beta&="+str(latex(fitEq(met.betaNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\tau&="+str(latex(fitEq(met.tauNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\pi&="+str(latex(fitEq(met.piNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\epsilon&="+str(latex(fitEq(met.epsilonNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\gamma&="+str(latex(fitEq(met.gammaNP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\newpage

\textbf{Relations between spin coefficients: }


%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\newpage
\textbf{Ricci tensor components: }
\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{00}&="+str(latex(fitEq(met.Phi00NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{01}&="+str(latex(fitEq(met.Phi02NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{10}&="+str(latex(fitEq(met.Phi10NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{11}&="+str(latex(fitEq(met.Phi11NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{12}&="+str(latex(fitEq(met.Phi12NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{20}&="+str(latex(fitEq(met.Phi20NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{21}&="+str(latex(fitEq(met.Phi21NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"\Phi_{22}&="+str(latex(fitEq(met.Phi22NP.expr()))))

tmpstr=r'''
\end{dmath}

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\newpage
\textbf{Weyl tensor components: }
\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\Psi_0&="+str(latex(fitEq(met.Psi0NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\Psi_1&="+str(latex(fitEq(met.Psi1NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\Psi_2&="+str(latex(fitEq(met.Psi2NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\Psi_3&="+str(latex(fitEq(met.Psi3NP.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}
'''
fout.write(tmpstr)
fout.write(r"\Psi_4&="+str(latex(fitEq(met.Psi4NP.expr()))))
tmpstr=r'''
\end{dmath}

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\newpage
\textbf{Newman-Penrose equations: }
\begin{dmath}'''
fout.write(tmpstr)

fout.write(r"NP1&="+str(latex(fitEq(met.NPeq1.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP2&="+str(latex(fitEq(met.NPeq2.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP3&="+str(latex(fitEq(met.NPeq3.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP4&="+str(latex(fitEq(met.NPeq4.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP5&="+str(latex(fitEq(met.NPeq5.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP6&="+str(latex(fitEq(met.NPeq6.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP7&="+str(latex(fitEq(met.NPeq7.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP8&="+str(latex(fitEq(met.NPeq8.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP9&="+str(latex(fitEq(met.NPeq9.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP10&="+str(latex(fitEq(met.NPeq10.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP11&="+str(latex(fitEq(met.NPeq11.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP12&="+str(latex(fitEq(met.NPeq12.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP13&="+str(latex(fitEq(met.NPeq13.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP14&="+str(latex(fitEq(met.NPeq14.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP15&="+str(latex(fitEq(met.NPeq15.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP16&="+str(latex(fitEq(met.NPeq16.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP17&="+str(latex(fitEq(met.NPeq17.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"NP18&="+str(latex(fitEq(met.NPeq18.expr()))))

tmpstr=r'''
\end{dmath}

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

\newpage
\textbf{Bianchi identities: }

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI1&="+str(latex(fitEq(met.BI1.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI2&="+str(latex(fitEq(met.BI2.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI3&="+str(latex(fitEq(met.BI3.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI4&="+str(latex(fitEq(met.BI4.expr())))+r' \\ '),
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI5&="+str(latex(fitEq(met.BI5.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI6&="+str(latex(fitEq(met.BI6.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI7&="+str(latex(fitEq(met.BI7.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI8&="+str(latex(fitEq(met.BI8.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI9&="+str(latex(fitEq(met.BI9.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI10&="+str(latex(fitEq(met.BI10.expr())))+r' \\ ')
tmpstr=r'''
\end{dmath}

\begin{dmath}'''
fout.write(tmpstr)
fout.write(r"BI11&="+str(latex(fitEq(met.BI11.expr()))))
tmpstr=r'''
\end{dmath}

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%


\newpage
\textbf{Petrov classification:} \\
\vspace{2cm}
\textit{Petrov class:}
\vspace{2cm}


'''
fout.write(tmpstr)
write_all_transformations(fout, met)
tmpstr = r'''

%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%


\begin{mdframed}
\textbf{Calculated by:}

\textit{Name: }...

\textit{E-mail:}...

\textit{Date:15.02.2025 }...
\end{mdframed}
\end{document}
'''
fout.write(tmpstr)
fout.close()
```
