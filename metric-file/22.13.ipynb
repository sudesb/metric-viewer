```python
reset()
from SageNP import NewmanPenrose
MyManifold = Manifold(4 , 'MyManifold', r'\mathcal{Man}')
MyCoordinates.<t,z,rho,phi> = MyManifold.chart(r't  z rho:\rho phi:\phi')


tetrad_type = 'covariant' 

var('B0,A')
A = 1 + (B0**2 * rho**2) / 4
sqrt2 = sqrt(2)

lvec = [ A/sqrt(2), A/sqrt(2), 0, 0 ]

nvec = [ A/sqrt(2), -A/sqrt(2), 0, 0 ]

mvec = [ 0, 0, A/sqrt(2), rho*A**-1/sqrt(2) ]

mbarvec = [ 0, 0, A/sqrt(2), -rho*A**-1/sqrt(2) ]


met = NewmanPenrose(MyManifold, MyCoordinates, lvec, nvec, mvec, mbarvec, tetrad_type)
f = MyManifold.scalar_field(function('f')(*MyCoordinates))

met.calculate_spincoefficients()
met.calculate_Weyl()
met.calculate_Ricci()
met.calculate_NPeq()
met.calculate_Bianchi()
met.Petrov_fromWeyl()

met.show_spincoefficients()
met.show_Weyl()
met.show_Ricci()
met.show_NPeq()
met.show_Bianchi()



def fitEq(expr):
    expanded_expr = expr.expand()

    latex_expr = latex(expanded_expr)

    return latex_expr
```
